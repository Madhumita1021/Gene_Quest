```python
import requests
import pandas as pd

# UBQLN2 gene region
region = "X:56563627-56567868"
url = f"https://rest.ensembl.org/overlap/region/human/{region}?feature=variation"
headers = {"Content-Type": "application/json"}

response = requests.get(url, headers=headers)

if response.status_code != 200:
    raise Exception("❌ Failed to fetch variant data from Ensembl")

data = response.json()
results = []

for item in data:
    var_id = item.get("id", "NA")
    consequence = item.get("most_severe_consequence", "NA")
    allele_freq = "NA"
    for colocated in item.get("colocated_variants", []):
        gnomad = colocated.get("gnomad", {})
        if "AF" in gnomad:
            allele_freq = gnomad["AF"]
            break

    results.append({
        "Variant ID": var_id,
        "Consequence": consequence,
        "Allele Frequency (gnomAD)": allele_freq
    })

# Create DataFrame
df = pd.DataFrame(results).drop_duplicates()

print(f"UBQLN2 gene is located at: Chromosome X, 56563627-56567868")
print(f"Number of variants found overlapping UBQLN2: {len(df)}")

# Filter to variants that have allele frequency data
df_with_af = df[df["Allele Frequency (gnomAD)"] != "NA"]

print(f"\nNumber of variants with gnomAD Allele Frequency data: {len(df_with_af)}")

# Display the filtered DataFrame - you can adjust .head() for how many rows to display
print("\nTop variants with Allele Frequency (gnomAD):")
print(df_with_af.head(10))
```
